```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the CSV files
csv_files = [ 'plif_results.csv', 'implif_results.csv']
dataframes = []

for file in csv_files:
    if os.path.exists(file):
        df = pd.read_csv(file)
        dataframes.append(df)
    else:
        print(f"File {file} not found.")

if not dataframes:
    print("No CSV files were loaded. Exiting.")
    exit()
```

```python
# Plotting
plt.figure(figsize=(12, 8))

# Training Loss
plt.subplot(2, 2, 1)
for i, df in enumerate(dataframes):
    plt.plot(df['epoch'], df['train_loss'], label=f'Model {i+1}')
plt.title('Training Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Training Accuracy
plt.subplot(2, 2, 2)
for i, df in enumerate(dataframes):
    plt.plot(df['epoch'], df['train_acc'], label=f'Model {i+1}')
plt.title('Training Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Validation Loss
plt.subplot(2, 2, 3)
for i, df in enumerate(dataframes):
    plt.plot(df['epoch'], df['val_loss'], label=f'Model {i+1}')
plt.title('Validation Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Validation Accuracy
plt.subplot(2, 2, 4)
for i, df in enumerate(dataframes):
    plt.plot(df['epoch'], df['val_acc'], label=f'Model {i+1}')
plt.title('Validation Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Time per Epoch
for i, df in enumerate(dataframes):
    plt.plot(df['epoch'], df['epoch_time'], label=f'Model {i+1}')
plt.title('Epoch Time vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Time (seconds)')
plt.legend()

plt.show()
```
